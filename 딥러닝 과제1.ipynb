```python
# Gradient Descent를 만들어보자

# 1. 편미분방정식 생성하기
# 2. 오차함수 생성하기
# 3. 실제 Gradient descent 사용하기

# 1. 편미분방정식의 입력은 str으로 받으며, 형식은 w1 * x**2 와 같은 형식으로 받는다.
# # 1) 먼저 space 기준으로 stack을 쌓아 계산하자.

# ex) fx = X**5  ==> derivative(fx) == 5 * x**4
# ex) fx = X**3 + 2*x**2 + 1  ==> derivative(fx) == 3 * (x**2) + 4*x


def derivative(fx, var_x):
    new_x = list(map(lambda x:x.strip(), fx.split('+')))
    
    for vx in new_x:
        oper_stack = []
        num_stack = []
        tmp = [idx if d=='*' else -1 for idx, d in enumerate(vx)]
        i = 0
        
        while i < len(tmp):
            if vx[tmp[i]] == '*':
                if tmp[i]+1 == tmp[i+1]:
                    oper_stack.append('**')
                    i += 2
                else:
                    oper_stack.append('*')
                    i += 1
            else:
                num_stack.append(vx[i])
                i += 1
        print(oper_stack, num_stack)
    
    return new_x

fx = 'X**3 + 2*x**2 + 1'
print(derivative(fx, 0))
```
